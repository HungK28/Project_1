```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests
import scipy.stats as st
from scipy.stats import linregress
from scipy.stats import sem
import numpy as np
import gmaps
```

```python
ppp_file = "../Resources/California_PPP.csv"
ppp_data = pd.read_csv(ppp_file)
ppp_data_df = pd.DataFrame(ppp_data)
ppp_data_df.head()
```

```python
ppp_data_df.describe()
```

```python
total_money = ppp_data_df["CurrentApprovalAmount"].sum()

```

```python
total_loans = ppp_data_df['ProcessingMethod'].count()
loan_type_count = ppp_data_df["ProcessingMethod"].value_counts()
loan_type_percentage = round(loan_type_count/ total_loans * 100, 2)

loan_demo_df = pd.DataFrame({"Percentage of loans": loan_type_percentage, "Total Count": loan_type_count})

loan_demo_df
```

```python
pie_plot = loan_type_count.plot.pie(autopct="%1.1f%%", title= "PPP vs PPS loans")
```

```python
bank_summary_df = ppp_data_df.groupby('OriginatingLender')

#bank_amount_mean = bank_summary_df['CurrentApprovalAmount'].mean()
bank_loans = bank_summary_df['CurrentApprovalAmount'].count()
bank_amount_total = bank_summary_df['CurrentApprovalAmount'].sum()

# Assemble the resulting series into a single summary dataframe.
bank_summary_amount_df = pd.DataFrame({
                                'Bank total amount loaned': bank_amount_total,
                                'Loan Count': bank_loans})
bank_summary_amount_df
```

```python
bank_summary_amount_df = bank_summary_amount_df.reset_index()
```

```python
sorted_bank = bank_summary_amount_df.sort_values("Bank total amount loaned", ascending=False).head(10)

sorted_bank
```

```python
bank_summary_amount_df.columns
```

```python
#plt.bar(sorted_bank["OriginatingLender"], sorted_bank["Loan Count"] )
#bar_plot = sorted_bank["OriginatingLender"].value_counts().plot.bar
plt.bar(sorted_bank["OriginatingLender"], sorted_bank["Bank total amount loaned"])
plt.xticks(sorted_bank["OriginatingLender"], rotation = "vertical")
plt.xlabel("Bank")
plt.ylabel("Bank Total Amount Loaned (Billions)")
plt.title("Amount per Bank")
plt.show()
```

```python
plt.bar(sorted_bank["OriginatingLender"], sorted_bank["Loan Count"])
plt.xticks(sorted_bank["OriginatingLender"], rotation = "vertical")
plt.xlabel("Bank")
plt.ylabel("Loan Count")
plt.title("Loans per Bank")
plt.show()
```
